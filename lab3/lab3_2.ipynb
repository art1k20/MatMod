\begin{equation*}  
 \begin{cases}
   0 \leq y \leq 1 \;\; if \;\; 0 \leq x \leq 1,
   \\
   0 \leq y \leq 3 \;\; if \;\; 1< x \leq 3,
   \\
   2 \leq y \leq 3 \;\; if \;\; 3< x \leq 4,
   \\
   
 \end{cases}
\end{equation*}

```python
import random as rnd
import matplotlib.pyplot as plt
```

```python
def rnd_points(x_h, x_l, y_h, y_l, sample_size):
    x = [rnd.uniform(x_l,x_h) for i in range(0,sample_size)]
    y = [rnd.uniform(y_l,y_h) for i in range(0,sample_size)]
    return x,y
```

```python
def calculate(x, y, sample_size):
    counter = 0
    inside_x = []
    inside_y = []
    outside_x = []
    outside_y = []
    for i in range(0,sample_size):
        if (x[i]<1 and y[i] >1) or (x[i] > 3 and y[i] < 2):
            outside_x.append(x[i])
            outside_y.append(y[i])
        else:
            counter += 1
            inside_x.append(x[i])
            inside_y.append([y[i]])
    return counter, inside_x,inside_y, outside_x, outside_y
```

```python
def area(x_l,x_h,y_l,y_h, sample_size, inside):
    return  (inside / sample_size) * (x_h - x_l) * (y_h - y_l)
```

```python
def print_scatter_plot(inside_x,inside_y, outside_x,outside_y):
    point_size = 2
    
    plt.scatter(inside_x, inside_y, c = 'blue', s = point_size)

    
    plt.scatter(outside_x, outside_y, c = 'yellow', s = point_size)

    plt.show()  
```

```python
sample_size = 10000
low_x, high_x = 0, 4
low_y, high_y = 0, 3

x,y = rnd_points(low_x, high_x, low_y, high_y, sample_size)
count, inside_x,inside_y, outside_x, outside_y = calculate(x,y,sample_size)

print('Площа фігури', area(low_x, high_x, low_y, high_y, sample_size, count))
print_scatter_plot(inside_x,inside_y, outside_x, outside_y)
```

```python
sample_size = 30000
low_x, high_x = 0, 4
low_y, high_y = 0, 3

x,y = rnd_points(low_x, high_x, low_y, high_y, sample_size)
count, inside_x,inside_y, outside_x, outside_y = calculate(x,y,sample_size)

print('Площа фігури', area(low_x, high_x, low_y, high_y, sample_size, count))
print_scatter_plot(inside_x,inside_y, outside_x, outside_y)
```
